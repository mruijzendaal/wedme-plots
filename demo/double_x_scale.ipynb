```python
import matplotlib.pyplot as plt
import numpy as np
import wedme

wedme.apply.paper()

plt.figure()

def plot_double_axis(ax1, ax2, x1, x2):
    forward = lambda x: np.interp(x, x1, x2)
    backward = lambda x: np.interp(x, x2, x1)

    xlims = [np.min(x1), np.max(x1)]

    ax1.set_xlim(*xlims)
    ax1.xaxis.set_tick_params(top=False)
    ax1.grid(True)

    ax2.set_xlim([forward(xi) for xi in xlims])
    ax2.grid(False)
    ax2.set_xscale(wedme.scale.CustomFuncScale(functions=[forward, backward][::-1], max_ticks=3))

ax1 = plt.gca()
ax2 = ax1.twiny()

x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100) ** 2
y  = x1

ax1.plot(x1, y)

plot_double_axis(ax1, ax2, x1, x2)

plt.show()
```
